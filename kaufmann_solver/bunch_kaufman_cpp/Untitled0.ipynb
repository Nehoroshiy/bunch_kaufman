```python
def csort(array, k):
    xxx = [0]*k
    for item in array:
        xxx[item] += 1
    sorted = []
    for i in xrange(k):
        for j in xrange(xxx[i]):
            sorted.append(i)
    return sorted
```

```python
array = [1, 4, 5, 6, 5, 4, 3, 2, 3, 4, 5, 6, 5, 4, 3, 2, 4, 2, 1, 1, 1, 2, 3, 4, 5, 6, 0, 0, 0, 0]
k = 7
```

```python
print csort(array, k)
```
